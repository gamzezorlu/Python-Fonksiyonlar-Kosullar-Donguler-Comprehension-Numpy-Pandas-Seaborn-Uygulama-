# Görev 18:  30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazınız.
# Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz. (apply ve lambda yapılarını kullanınız)

```python
def age_30(age):
    if age<30:
        return 1
    else:
        return 0
```

```python
df["age_flag"] = df["age"].apply(lambda x : age_30(x))
```

```python
df["age_flag"] = df["age"].apply(lambda x: 1 if x<30 else 0)
```

# Görev 19: Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

# Görev 20: Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill  değerlerinin toplamını, min, max ve ortalamasını bulunuz.

```python
df.groupby("time").agg({"total_bill": ["sum","min","mean","max"]})
```

# Görev 21: Günlere ve time göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.

```python
df.groupby(["day","time"]).agg({"total_bill": ["sum","min","mean","max"]})
```

# Görev 22:Lunch zamanına ve kadın müşterilere ait total_bill ve tip  değerlerinin day'e göre toplamını, min, max ve ortalamasını bulunuz.

```python
df[(df["time"] == "Lunch") & (df["sex"] == "Female")].groupby("day").agg({"total_bill": ["sum","min","max","mean"],
                                                                           "tip":  ["sum","min","max","mean"]})
```

# Görev 23: size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir?

```python
df.loc[(df["size"] < 3) & (df["total_bill"] >10 ) , "total_bill"].mean()
```

# Görev 24: total_bill_tip_sum adında yeni bir değişken oluşturun. Her bir müşterinin ödediği totalbill ve tip in toplamını versin.

```python
df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
df.head()
```

# Görev 25: total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.

```python
new_df = df.sort_values("total_bill_tip_sum", ascending=False)[:30]
new_df.shape
```
